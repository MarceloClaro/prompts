```python
import os
from crewai import Agent, Task, Crew, Process
from langchain_openai import ChatOpenAI

# Configurando o LLM padrão para todos os agentes e tarefas
OPENAI_API_KEY = "Sua Chave"
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
os.environ["OPENAI_MODEL_NAME"] = 'gpt-3.5-turbo'

chat = ChatOpenAI(api_key=OPENAI_API_KEY)

# Quantidade de agentes e rodadas fornecida pelo usuário
num_agentes = 4  # Exemplo: 3 agentes
num_rodadas = 3  # Exemplo: 5 rodadas



# Criando debatedores
debatedores = []
for i in range(1, num_agentes + 1):
    debatedor = Agent(
        role=f'Debatedor {i}',
        goal='Propor e aprimorar soluções para o problema proposto. Todas as saídas devem ser em português do Brasil',
        verbose=True,
        memory=True,
        backstory=(
            f'Você é o Debatedor {i}. Sua tarefa é propor soluções inovadoras para o problema '
            'e aprimorar suas ideias baseadas nas soluções dos outros debatedores. v'
        ),
        allow_delegation=False
    )

    debatedores.append(debatedor)

# Criando agente gerente
gerente = Agent(
    role='Gerente',
    goal='Coordenar os debatedores e consolidar a solução final. Todas as saídas devem ser em português do Brasil',
    verbose=True,
    memory=True,
    backstory=(
        'Você é o Gerente. Sua tarefa é coordenar os debatedores, garantir que todas as rodadas de debate sejam realizadas '
        'e consolidar a solução final para o problema. Todas as saídas devem ser em português do Brasil'
    ),
    allow_delegation=True
)

# Criando tarefas de debate
tarefas_debate = []
for rodada in range(num_rodadas):
    for i in range(1, num_agentes + 1):
        tarefa = Task(
            description=(
                f'Rodada {rodada + 1}: Debatedor {i}, apresente sua solução parcial para o problema. '
                'Depois, leia e reflita sobre as soluções dos outros debatedores. Atualize sua solução com base nessas reflexões. Todas as saídas devem ser em português do Brasil'
            ),
            expected_output='Uma solução atualizada para o problema. Todas as saídas devem ser em português do Brasil',
            agent=debatedores[i-1],
            async_execution=False,
            output_file=f'solucao_debatedor_{i}_rodada_{rodada + 1}.txt'
        )
        tarefas_debate.append(tarefa)

# Tarefa final de consolidação da solução
tarefa_consolidacao = Task(
    description=(
        'Leia todas as soluções finais dos debatedores e consolide uma solução final abrangente para o problema. Todas as saídas devem ser em português do Brasil'
    ),
    expected_output='Uma solução final consolidada para o problema. Todas as saídas devem ser em português do Brasil',
    agent=gerente,
    async_execution=False,
    output_file='solucao_final.txt'
)

# Formando a crew de debate
crew_debate = Crew(
    agents=debatedores + [gerente],
    tasks=tarefas_debate + [tarefa_consolidacao],
    process=Process.sequential  # Execução sequencial
)

# Iniciando o debate
resultado_final = crew_debate.kickoff(inputs={'problema': 'Como recuperar casas que foram destruidas nas enchentes do Rio Grande do Sul.'})
print(resultado_final)
```
